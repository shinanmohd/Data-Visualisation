### deaths over cases in first 20 countries

```python
first20=country.sum(numeric_only=True).head(20)
```

```python
first20
```

```python
a=first20[['cases','deaths']]
```

```python
plt.plot(a['cases'],a['deaths'],marker='d',c='g')
plt.xlabel('cases')
plt.ylabel('deaths')
plt.title('deaths over cases');
```

```python
plt.bar(first20['cases'],first20['deaths'],color='red',edgecolor='black')
;
```

```python
a=first20[['cases','deaths']]
a
```

```python
a.plot(kind='line',colormap='plasma',marker='d')
plt.xlabel('cases')
plt.ylabel('deaths')
plt.title('deaths over cases');
```

```python
a.plot(kind='bar',colormap='viridis',edgecolor='black')
plt.xlabel('cases')
plt.ylabel('deaths')
plt.title('deaths over cases');
```

```python
a.plot(kind='hist',colormap='magma')
plt.xlabel('cases')
plt.ylabel('deaths')
plt.title('deaths over cases');
```

```python
a.hist(edgecolor='black')
plt.suptitle('cases and deaths rate');
```

```python
plt.boxplot([a['cases'],a['deaths']]);
```

```python
a.boxplot(labels=a.columns)
plt.xlabel('cases')
plt.ylabel('deaths')
plt.title('deaths over cases');
```

```python
plt.violinplot([a['cases'],a['deaths']])
plt.xlabel('cases')
plt.ylabel('deaths')
plt.title('deaths over cases');
```
