```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fortu=pd.read_csv(r'C:\Users\hp\Data Analyst\csv\fortune1000.csv')
```

```python
fort=fortu.head(10)
```

```python
plt.figure(figure=(10,5))
plt.plot(fort['Company'],fort['Revenues'],marker='d',c='r')
plt.plot(fort['Company'],fort['Profits'],marker='o',c='g')
plt.xlabel('Company')
plt.ylabel('P or R')
plt.legend(['revenue','profit'])
plt.xticks(rotation=90)
plt.title('Data');
```

```python
plt.bar(fort['Company'],fort['Revenues'],color='green')
plt.xticks(rotation=90);
```

```python
plt.hist(fort['Profits'],bins=5,edgecolor='black')
plt.xlabel('P or R')
plt.title('Data');
```
