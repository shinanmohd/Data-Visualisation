## Description

## Heatmap representation

```python
studperf=studentperformance.select_dtypes(include='number')
```

```python
sns.heatmap(studperf.corr(),annot=True)
```

## representing counts of each group using countplot

```python
sns.countplot(x='race/ethnicity',data=studentperformance);
```

## student performane representation using lineplot

```python
sns.lineplot(x='reading score',y='writing score',data=studentperformance,marker='s')
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x='reading score',y='writing score',hue='test preparation course',data=studentperformance)
plt.title('Student Performance');
```

## representing writing score of each group with the highest qualification each achieved

```python
plt.figure(figsize=(10,8))
sns.barplot(x='race/ethnicity',y='writing score',hue='parental level of education',data=studentperformance)
plt.title('writing score of each group with the highest qualification');
```

## checking if their any outliers using boxplot

```python
sns.boxplot(studentperformance)
```

## comparing students reading score with writing score using scatter plot

```python
sns.scatterplot(x='reading score',y='writing score',data=studentperformance,hue='race/ethnicity');
```

## which group has a high score in maths using violineplot

```python
sns.violinplot(x='race/ethnicity',y='math score',data=studentperformance);
```

## pairplot

```python
sns.pairplot(studentperformance);
```

## jionplot

```python
sns.jointplot(x='reading score',y='writing score',data=studentperformance,kind='scatter');
```

## facetgrid

```python
f=sns.FacetGrid(data=studentperformance,row='race/ethnicity',col='parental level of education')
f.map(sns.kdeplot,'math score')
plt.tight_layout()
plt.show()
```

## piechart reprensting which group has high score in maths

```python
studentperformance.groupby('race/ethnicity')['math score'].sum().plot(kind='pie',explode=(0,0,.1,0,0),shadow=True,autopct="%1.2f%%");
```

```python
sns.histplot(x='race/ethnicity',data=studentperformance);
```
